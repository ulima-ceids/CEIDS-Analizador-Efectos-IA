```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
audio_path = "../data/raw/test.wav"

audio, sr = librosa.load(audio_path, sr=44100)

print("Duración del audio (segundos):", len(audio) / sr)
print("Frecuencia de muestreo:", sr)

plt.figure(figsize=(12, 3))
librosa.display.waveshow(audio, sr=sr)
plt.title("Waveform de test.wav")
plt.show()
```

```python
stft = librosa.stft(audio)
spectrogram = librosa.amplitude_to_db(np.abs(stft), ref=np.max)

plt.figure(figsize=(12, 5))
librosa.display.specshow(spectrogram, sr=sr, x_axis="time", y_axis="hz", cmap="magma")
plt.colorbar(format="%+2.0f dB")
plt.title("espectrograma")
plt.show()
```
